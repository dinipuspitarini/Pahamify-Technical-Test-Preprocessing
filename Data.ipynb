```python
df = pd.read_csv('Pahamify Data - transaction_transaction.csv')
df.head()
```

```python
df.info()
```

```python
df1 = df[df['TransactionStatus'] == "settlement"]
df1
```

```python
df1 = df1[["CompletedOrderTransactionDate", "GrossPrice"]]
df1['ExtractDate'] = df1['CompletedOrderTransactionDate'].str.extract(r'(.*-..)')
df1['Month'] = df1['ExtractDate']
df1['Month'] = df1['Month'].str.replace(r'[0-9]{4}-*', '', regex=True)
df1['Month'] = df1['Month'].str.replace(r'-[0-9][0-9]', '', regex=True)
df1['Month'] = df1['Month'].map({'07': 'Juli 2019', '08': 'Agustus 2019', '09': 'September 2019', '10': 'Oktober 2019',
                                 '11': 'November 2019', '12': 'Desember 2019', '01': 'Januari 2020'})
cat_month = CategoricalDtype(['Juli 2019', 'Agustus 2019', 'September 2019', 'Oktober 2019', 'November 2019', 'Desember 2019', 'Januari 2020'], 
    ordered=True)
df1['Month'] = df1['Month'].astype(cat_month)
df1.sort_values(by=['ExtractDate'], inplace = True, ascending = True)
df1.drop(columns='CompletedOrderTransactionDate', axis = 1, inplace=True)
df1 = df1.groupby(['Month'], as_index=False).agg({'GrossPrice': ['sum']})
df1.columns = ['Month', 'Pendapatan']
df1
```

```python
df1.to_csv('MRR.csv', index = False)
```

```python
data = pd.read_csv('Pahamify Data - user_data_users.csv')
data.head()
```

```python
data['ExtractDate'] = data['RegistrationDate'].str.extract(r'(.*-..)')
data['ExtractDate'] = pd.to_datetime(data['ExtractDate'])
data.sort_values(by=['ExtractDate'], inplace = True, ascending = True)
data.head()
```

```python
data = data.groupby(['ExtractDate'], as_index=False).agg({'Email': ['count']})
data.columns = ['Date', 'Registered User']
data['Total User'] = data['Registered User'].cumsum(axis = 0)
data
```

```python
data.to_csv('total user.csv', index = False)
```
